```python
import pandas as pd

df = pd.read_csv('data/renwu.csv')
```

```python
# Nan替换
df.loc[df.年代列表.isna(),'年代列表'] = '未知'
df.loc[df.字号.isna(),'字号'] = '未知'
```

```python
df.isna().any()
```

```python
#去掉多余文字
def get_cata(value):
    return value.replace('主效','').replace('势力','')

df.主效 = df.主效.map(get_cata)
```

```python
df.主效.value_counts()
```

```python
#添加效力数
def get_cata_list_len(value):
    return len([x for x in value.split('|') if len(x)>0])
df.效力 = df.效力.map(get_cata)
df['效力数'] = df.效力.map(get_cata_list_len)
```

```python
#去掉总章节多余文字
def get_total_chapter(value):
    res = value[1:-1]
    if len(res) > 0:
        return int(res)
    else :
        return 0
df.总章节 = df.总章节.map(get_total_chapter)
```

```python
# 查看效力数分布，效力数最多的人是刘备，各种叛变。
df.效力数.value_counts()
```

```python
df.总章节.value_counts()
```

```python
df.配偶.value_counts()
```

```python
#添加配偶数，最多的两个人，曹操曹丕
def get_mate_len(value):
    if value == '未知':
        return 0
    else:
        res = [x for x in value.split('|') if len(x)>0]
        return len(res)
    
df['配偶数'] = df.配偶.map(get_mate_len)
```

```python
df.配偶数.value_counts()
```

```python
#去掉总年代数多余文字
df.总年数 = df.总年数.map(get_total_chapter)
```

```python
#去掉字号多余文字
def process_zihao(value):
    if value[0] == '字':
        return value[1:]
    else :
        return value
df.字号 = df.字号.map(process_zihao)
df.head()
```

```python
from CrawlNovel import load_pkl
novel = load_pkl()
novel[1]
```

```python

def get_chuchang_len(value):
    res = 0
    for n in novel:
        if value.名字 in n['content'] or value.字号 in n['content']:
            
            res += n['content'].count(value.名字)
            res += n['content'].count(value.字号)
    return res

df['出场次数'] = df.apply(get_chuchang_len,axis=1)
```

```python
df.to_csv('data/renwu.csv',index=False)
```

```python
rela_list = []
def mate_rela(value):
    name = value['名字']
    mates = value['配偶'].split("|")
    for mate in mates:
        rela_list.append([value['编号'],name,'mate',mate])
        
def cata_rela(value):
    name = value['名字']
    cateds = value['效力'].split("|")
    for cata in cateds:
        if cata == value['主效']:
            rela_list.append([value['编号'],name,'cata',cata])
        else:
            rela_list.append([value['编号'],name,'cated',cata])
            
def brother_rela(value):
    name = value['名字']
    brothers = value['兄弟'].split("|")
    for brother in brothers:
        rela_list.append([value['编号'],name,'brother',brother])

def father_rela(value):
    name = value['名字']
    father = value['父亲']
    rela_list.append([value['编号'],name,'father',father])
    
def mother_rela(value):
    name = value['名字']
    mother = value['母亲']
    rela_list.append([value['编号'],name,'mother',mother])
    
df.apply(mate_rela,axis=1)
df.apply(cata_rela,axis=1)
df.apply(brother_rela,axis=1)
df.apply(father_rela,axis=1)
df.apply(mother_rela,axis=1)

rela_colume = ['编号','实体1','关系','实体2']

rela_df = pd.DataFrame(columns=rela_colume,data=rela_list)
rela_df.to_csv('data/renwu_relate.csv',index=False)
```
