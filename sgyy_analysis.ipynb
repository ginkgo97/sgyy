```python
def get_cata(value):
    return value.replace('主效','').replace('势力','')

df.主效 = df.主效.map(get_cata)
```

```python
df.主效.value_counts()
```

```python
def get_cata_list_len(value):
    return len([x for x in value.split('|') if len(x)>0])
df.效力 = df.效力.map(get_cata)
df['效力数'] = df.效力.map(get_cata_list_len)
```

```python
def get_total_chapter(value):
    res = value[1:-1]
    if len(res) > 0:
        return int(res)
    else :
        return 0
df.总章节 = df.总章节.map(get_total_chapter)
```

```python
def get_mate_len(value):
    if value == '未知':
        return 0
    else:
        res = [x for x in value.split('|') if len(x)>0]
        return len(res)
    
df['配偶数'] = df.配偶.map(get_mate_len)
```

```python
def process_zihao(value):
    if value[0] == '字':
        return value[1:]
    else :
        return value
df.字号 = df.字号.map(process_zihao)
df.head()
```

```python
from CrawlNovel import load_pkl
novel = load_pkl()
novel[1]
```

```python

def get_chuchang_len(value):
    res = 0
    for n in novel:
        if value.名字 in n['content'] or value.字号 in n['content']:
            
            res += n['content'].count(value.名字)
            res += n['content'].count(value.字号)
    return res

df['出场次数'] = df.apply(get_chuchang_len,axis=1)
```
